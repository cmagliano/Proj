The purpose of this code is to capture image files found in the rows of an Excel file and save them to Google Drive's MyDrive.

# Selecting Artists
After obtaining the Excel file, the artwork of 4 artists is separated into another dataframe.

```python
df_artpicies = df_catalog[df_catalog['AUTHOR'].isin(['MONET, Claude','BOTTICELLI, Sandro','RENOIR, Pierre-Auguste','GOGH, Vincent van'])]
df_artpicies
```

#Changing URL and saving the files in My Drive
Changing the HTML URL to get the files and move them to Google Drive

```python
gdrive_link = "/content/drive/MyDrive/Artpicies/"

i = 0
author = " "
for row in df_artpicies.itertuples():
    url = row.URL

    new_url = regex.sub(
    '(.html)',
    '.jpg',
    regex.sub('(/html)', '/art', url))

    if (row.AUTHOR.split(',')[0] == author) or author == " ":
      i+=1
    else:
      i=1

    author = row.AUTHOR.split(',')[0]

    filename =author +"_"+ str(i)+ '.jpg'
    urllib.request.urlretrieve(new_url, filename)
    shutil.copy(filename, gdrive_link)
```
