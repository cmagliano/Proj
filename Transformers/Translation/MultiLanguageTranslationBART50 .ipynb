```python
from transformers import MBartForConditionalGeneration, MBart50Tokenizer
```

```python
def download_model():
  model_name = 'facebook/mbart-large-50-many-to-many-mmt'
  model=MBartForConditionalGeneration.from_pretrained(model_name)
  tokenizer = MBart50Tokenizer.from_pretrained(model_name)
  return model, tokenizer
```

```python
model, tokenizer = download_model()
```

```python
text = "Eu sou Claudia e estou estudando ciência de dados"
tokenizer.src_lang = 'pt_XX'
encoded_port_text = tokenizer(text, return_tensors='pt')
#generated_tokens = model.generate(**encoded_port_text, forced_bos_token_id=tokenizer.lang_code_to_id["it_IT"])
generated_tokens = model.generate(**encoded_port_text, forced_bos_token_id=tokenizer.lang_code_to_id["en_XX"])
out=tokenizer.batch_decode(generated_tokens,skip_special_tokens=True)
out
```
